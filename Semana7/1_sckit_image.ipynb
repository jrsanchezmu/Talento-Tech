```python
import skimage
# Carga imagen de skimage.data
camera = skimage.data.camera()
# Explora imagen
print(type(camera))
print(camera.shape)

# Mostrar la imagen

skimage.io.imshow(camera)
skimage.io.show()
```

```python
brick = skimage.data.brick()
# Explora imagen
print(type(brick))
print(brick.shape)

# Mostrar la imagen

skimage.io.imshow(brick)
skimage.io.show()
```

```python
# Ruta de la imagen
image_path = 'imagen.jpg'

# Cargar la imagen
image = skimage.io.imread(image_path)

# Mostrar la imagen
skimage.io.imshow(image)
skimage.io.show()
```

```python
# Rotar la imagen en sentido antihorario
rotated_image = skimage.transform.rotate(image, angle=30)

# Mostrar la imagen rotada
skimage.io.imshow(rotated_image)
skimage.io.show()
```

```python
# Aplicar un filtro Gaussiano para suavizar la imagen
smoothed_image = skimage.filters.gaussian(image, sigma=0.5)

# Mostrar la imagen suavizada
skimage.io.imshow(smoothed_image)
skimage.io.show()
```
